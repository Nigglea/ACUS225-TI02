```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
%matplotlib notebook
import numpy as np
import pandas as pd
from loudspeaker_box import main, open_box, utils
import IPython.display as disp
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
sns.set()
xticks = [2,4,8,16,31.5, 63, 125, 250, 500, 1000,2000,4000 ,8000]
```

## Carga de Dataset

```python
dataset = main.dataset()
df = dataset.loc['Eminence','DELTA_12A']
```

```python
bass_r = main.bass_reflex(df)
bass_r.parametros_TS()
bass_r.dimensiones()
bass_r.port()
bass_r.impedancia()
bass_r.NPS_1m()
```

```python
w   = bass_r.w
NPS = bass_r.NPS
ZE  = bass_r.mag_Ze
```

```python
fig,ax = plt.subplots(1,2,figsize=(17,7),tight_layout=True)
ax[0].set_title('Impedancia Electrica $Z_e$',fontsize=18)
ax[0].plot(w/(2*np.pi),ZE)
ax[0].set_xscale('log')
ax[0].set_xlabel('Hz',fontsize=14)
ax[0].set_ylabel('Ohm',fontsize=14)
ax[0].set_xlim([1,10025]);
ax[0].set_xticks(xticks);
ax[0].set_xticklabels(xticks,fontsize=14);

ax[1].plot(w/(2*np.pi),NPS)
ax[1].set_xscale('log')
ax[1].set_xlim([1,10025]);
ax[1].set_title('NPS',fontsize=18)
ax[1].set_xlabel('Hz',fontsize=14)
ax[1].set_ylabel('dB',fontsize=14)
ax[1].set_xticks(xticks);
ax[1].set_xticklabels(xticks,fontsize=14);
```
